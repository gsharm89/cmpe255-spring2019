```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr
```

```python
x = [17.5, 22, 29.5, 44.5, 64.5, 80]
y = [38, 36, 24, 20, 18, 28]
```

```python
plt.plot(x,y, "b.")
plt.xlabel("Age", fontsize=18)
plt.ylabel("# of deaths per 100,000", rotation=90, fontsize=18)
plt.axis([0, 100, 0, 40])
plt.show()
```

```python
X_b = np.c_[np.ones((6, 1)), x]  # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_new = np.array([[0], [100]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(x, y, "b.")
plt.axis([0, 100, 0, 40])
plt.ylabel("# of deaths per 100,000", rotation=90, fontsize=18)
plt.show()
```

```python
#predicting # of deaths at age 40
X_new = np.array([40])
X_new_b = np.c_[np.ones((1, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
print("The number of predicted deaths at age 40 are ", y_predict)
```

```python
#predicting # of deaths at age 60
X_new = np.array([40])
X_new_b = np.c_[np.ones((1, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
print("The number of predicted deaths at age 60 are ", y_predict)
```

```python
print('The correlation coefficient is : ', pearsonr(x,y))
```
